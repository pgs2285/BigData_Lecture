```python
import numpy as np
import pandas as pd
PREVIOUS_MAX_ROWS = pd.options.display.max_rows #display 할때 row몇개 불러올까
pd.options.display.max_rows = 20 #앞에꺼 10개 뒤에꺼 10개
np.random.seed(12345)
import matplotlib.pyplot as plt
import matplotlib
plt.rc('figure', figsize=(10, 6)) #rc로 크기 지정가능
np.set_printoptions(precision=4, suppress=True)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
#가장최근의 피겨와 서브플랏에 그린다
plt.plot(np.random.randn(50).cumsum(), 'k--')
ax2.hist(np.random.randn(100), bins=100, color='k', alpha=0.3)
ax1.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
plt.subplots_adjust(wspace=1, hspace =0)# 간격지정하기 (left, right, top, bottom. wspace,hspace 가있다.)
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=10, color='k', alpha=0.9, label=i**j) #alpha값은 투명도 조정
        plt.legend(loc='best')  
      
```

```python
plt.plot(np.random.randn(30).cumsum(), 'ks:',label='Test Data')
plt.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax2 = fig.add_subplot(2,2,2)
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                            rotation=90, fontsize='large')

ax.set_title('test plot')
ax.set_xlabel('number')
ax.plot(np.random.randn(1000).cumsum())
ax2.plot(np.arange(10))
```

```python
from numpy.random import randn
fig = plt.figure(); ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), 'k', label = 'one')
ax.plot(randn(1000).cumsum(), 'k', label = 'two')
ax.plot(randn(1000).cumsum(), 'k', label = 'three')
ax.legend(loc='best')
```

```python
from datetime import datetime
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
data = pd.read_csv('./examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
spx.plot(ax=ax, style='k-')
crisis_data=[
    (datetime(2007,10,11), 'Peak of bull market'),
    (datetime(2008,3,12), 'Bear Stearns Falls'),
    (datetime(2008,9,15), 'Lehman Bankrupty')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+75),
               xytext=(date,spx.asof(date)+225),
               arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
               horizontalalignment='left', verticalalignment='top')
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600,1800])
```

```python
tips=pd.read_csv('./examples/tips.csv')
party_counts = pd.crosstab(tips['day'],tips['size'])
```

```python
party_counts = party_counts.loc[:,2:5]
party_counts
```

```python
party_pcts=party_counts.div(party_counts.sum(1),axis =0)
party_pcts.plot.bar()
```

```python
macro = pd.read_csv('examples/macrodata.csv')
data = macro[['cpi','m1','tbilrate','unemp']]
# s_data = data[:5]
trans_data = np.log(data).diff().dropna()
# print(trans_data.head(),'\n', s_data)
trans_data
```

```python
sns.regplot('m1', 'unemp', data=trans_data) #선형 회귀 보여주는것
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha':0.2})
```
