마스크를 이용한 영상의 합성

```python
cv2.copyTo?
```

```python
src = cv2.imread('ch04/airplane.bmp', cv2.IMREAD_COLOR)
mask = cv2.imread('ch04/mask_plane.bmp', cv2.IMREAD_GRAYSCALE)
dst = cv2.imread('ch04/field.bmp', cv2.IMREAD_COLOR)

cv2.copyTo(src, mask, dst)
# dst[mask>0] = src[mask>0]

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('mask', mask)
cv2.waitKey()
cv2.destroyAllWindows()
```

cv2.TickMeter()를 활용한 성능측정

```python
src = cv2.imread('ch04/airplane.bmp', cv2.IMREAD_COLOR)
tm = cv2.TickMeter()
tm.start()
src = ~src
# src = 255 - src
tm.stop()

cv2.imshow('src', src)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(f'{tm.getTimeMilli():.3f} milli seconds')
```

## 영상의 정규화 및 확인

```python
src = cv2.imread('ch04/lenna.bmp', cv2.IMREAD_GRAYSCALE)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(src)
print(f'{minVal}, {maxVal}, {minLoc}, {maxLoc}')
```

```python
dst = cv2.normalize(src, None, 0, 255,cv2.NORM_MINMAX)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(dst)
print(f'{minVal}, {maxVal}, {minLoc}, {maxLoc}')
```

```python
dst2 = cv2.normalize(src, None, 100, 200,cv2.NORM_MINMAX)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(dst2)
print(f'{minVal}, {maxVal}, {minLoc}, {maxLoc}')
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('dst2', dst2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('ch05/hawkes.bmp', cv2.IMREAD_GRAYSCALE)
dst = cv2.equalizeHist(src)
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)
for i, img in enumerate([src, dst]):
    ax[i].imshow(img, cmap='gray')
```

```python
bgr = cv2.imread('ch05/lenna.bmp', cv2.IMREAD_COLOR)
YCrCb = cv2.cvtColor(bgr, cv2.COLOR_BGR2YCrCb)
Y, Cr, Cb = cv2.split(YCrCb)
Y2 = cv2.equalizeHist(Y)
Y2CrCb = cv2.merge([Y2, Cr, Cb])
bgr2 = cv2.cvtColor(Y2CrCb, cv2.COLOR_YCrCb2BGR)

import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)
for i, img in enumerate([bgr, bgr2]):
    ax[i].imshow(img[:,:,::-1])
```

히스토그램 역투영

```python
cv2.selectROI?
```

```python
import cv2
import numpy as np

src = cv2.imread('data/fruits.jpg')
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
hue, saturation, value = cv2.split(hsv)

roi = cv2.selectROI(src)
x, y, w, h = roi
roi_h = hue[y:y+h, x:x+w]

hist = cv2.calcHist([roi_h], [0], None, [64], [0, 256])
backP = cv2.calcBackProject([hue.astype(np.float32)], [0], hist, 
                            [0, 256], scale=1.0)
cv2.imshow('src', src)
cv2.imshow('backP', backP)
cv2.waitKey()
cv2.destroyAllWindows()
```

어파인 변환

```python
import cv2
src = cv2.imread('ch05/lenna.bmp')
M = cv2.getRotationMatrix2D((256, 256), 45, 0.5)
dst = cv2.warpAffine(src, M, dsize=(512,512))
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
