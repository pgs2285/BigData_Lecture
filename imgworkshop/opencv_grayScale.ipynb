```python
import cv2
from glob import glob
```

```python
gorani = glob('./Hydropotes inermis_half/*')
boar = glob('./wild boar2/*')
gorani
```

```python
for idx, i in enumerate(boar):
    img = cv2.imread(i)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_img  = cv2.resize(gray_img, dsize=(380, 380), interpolation=cv2.INTER_LINEAR)
    cv2.imwrite("./gray_boar/boar"+str(idx)+'.jpg', gray_img)
    print("process = " + str(idx))
print('done')    
```
